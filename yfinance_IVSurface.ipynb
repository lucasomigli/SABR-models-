```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import pandas as pd

import yfinance as yf
aapl = yf.Ticker("AAPL")

maturities = aapl.options

df = pd.concat([aapl.option_chain(x).calls for x in maturities])
df['contractSymbol'] = df['contractSymbol'].str.replace('AAPL', '').map(lambda x: str(x)[:-9])
df['contractSymbol'] = pd.to_numeric(pd.to_datetime(df['contractSymbol'], format='%y%m%d'))
```

```python
fig = plt.figure(figsize=(11, 8), dpi=80)
ax = fig.gca(projection='3d')
# I got the graph using trisurf 
graf = ax.plot(df["contractSymbol"], df["strike"].loc[15:50], df["impliedVolatility"])

ax.set_xlabel('Strikes')
ax.set_ylabel('Maturities')
ax.set_zlabel('Implied Volatility')

plt.show()
```

```python
fig = plt.figure(figsize=(11, 8), dpi=80)
ax = fig.gca()

for x in maturities:
    strikes = aapl.option_chain(x).calls['strike'].loc[15:50]
    IV = aapl.option_chain(x).calls['impliedVolatility'].loc[15:50]

    ax.plot(strikes, IV)

ax.set_xlabel('Strikes')
ax.set_ylabel('Implied Volatility')

plt.show()
```
